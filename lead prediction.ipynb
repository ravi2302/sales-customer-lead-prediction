```python
# Importing the dataset
dataset = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/train_wn75k28.csv')
dataset.isnull().sum()


```

```python
dataset['products_purchased'].fillna("0", inplace=True)
dataset['signup_date'].fillna("not signed up", inplace=True)
```

```python
dataset.isnull().sum()
```

```python
dataset.drop(columns="created_at",axis=1,inplace=True)
dataset.drop(columns="signup_date",axis=1,inplace=True)
dataset
```

```python
x_train = dataset.iloc[:,1:-1].values
y_train = dataset.iloc[:,-1]

```

```python
dataset2 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/test_Wf7sxXF.csv')
dataset2.isnull().sum()
```

```python
dataset2['products_purchased'].fillna("0", inplace=True)
dataset2['signup_date'].fillna("not signed up", inplace=True)
```

```python
dataset2.isnull().sum()
```

```python
dataset2.drop(columns="created_at",axis=1,inplace=True)
dataset2.drop(columns="signup_date",axis=1,inplace=True)
dataset2
```

```python
#importing the dataset
x_test = dataset2.iloc[:,1:].values
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
# Training the Decision Tree Classification model on the Training set
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
classifier.fit(x_train, y_train)
```

```python
# Predicting the Test set results
y_pred = classifier.predict(x_test)
df2 = pd.DataFrame(y_pred, columns = ['result'])
df1 = dataset2.iloc[:,0]
result = pd.concat([df1, df2], axis=1, join='inner')
display(result)
result.to_csv('/content/drive/MyDrive/Colab Notebooks/result.csv', index = False)
```
